```
import requests
from bs4 import BeautifulSoup

#2019ë…„ì˜ url listë¥¼ ë§Œë“¤ì!

url = "https://mini.snu.kr/index.php/cafe/set/2019-"
url_list= []
day31 = [1, 3, 5, 7, 8, 10, 12]
day30 = [4, 6, 9, 11]
for i in range(1, 13):
    if i in day31:
        for to31 in range(1, 32):
            url_list.append(url + str(i) + '-' + str(to31) + '/abcf')
    elif i in day30:
        for to30 in range(1, 31):
            url_list.append(url + str(i) + '-' + str(to30) + '/abcf')
    elif i==2:
        for to28 in range(1, 29):
            url_list.append(url + str(i) + '-' + str(to28) + '/abcf')

url_list
```

```
menu2019 = []
# 2019ë…„ì˜ ë©”ë‰´ ëª©ë¡ì„ í¬ë¡¤ë§í•´ì˜¤ì! (ë‹¨, ì˜ˆì™¸ë“¤ì„ ì œê±°)

for url in url_list:
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')

    menu = soup.find_all('td', {'class' : 'menu'})
    menu_list = []
    for a in menu:
        menu_list.append(a.get_text())
    for b in menu_list:
        if 'íœ´ë¬´' in b or 'íœ´ì—…'in b or 'íœ´ì ' in b or 'ê³µíœ´ì¼' in b or 'ê°œêµê¸°ë…ì¼' in b or 'ì½”ë¡œë‚˜'in b or 'ê³µê°„'in b or 'â—' in b or 'ëˆ„í…”ë¼ì™€í”Œ' in b or 'ìœ ë‹ˆì§œì¥' in b or 'ëŠ¥ì´ë²„ì„¯' in b or 'êµì§ì›ì‹ë‹¹' in b or 'ê¸‰ì‹' in b:
            None
        else:
          menu2019.append(b)

menu2019
```

```
# ì •ê·œí‘œí˜„ì‹ì„ ì¨ì„œ ê¹”ë”í•˜ê²Œ ë©”ë‰´ëª…ë§Œ ê°€ì ¸ì˜¤ê³ , ë¹ˆë„ë¥¼ ì•Œì•„ë³´ëŠ” counterí•¨ìˆ˜ ì‚¬ìš©

import re

my_str = ' '.join(menu2019)

p = re.compile('[ê°€-í£]+')  # \&?\s?[ê°€-í£]+
m = p.findall(my_str)

from collections import Counter

result = Counter(m)
print(result)
```

```
# counter ë”•ì…”ë„ˆë¦¬ì—ì„œ ê°ê° ë©”ë‰´, ë¹ˆë„ìˆ˜ì¸ key, value ë³€ìˆ˜ì— ì €ì¥

a1= result.keys()
a2 = result.values()
```

```
# seaborn ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìœ„í•´ íŒë‹¤ìŠ¤ ë°ì´í„° í”„ë ˆì„ìœ¼ë¡œ ê°’ ì €ì¥

import pandas as pd
import numpy as np

df=pd.DataFrame([x for x in zip(a1, a2)])
df
df1 = df.sort_values(by= 1, ascending = False)  #ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë¦¬

df2 = df1.head(15)                              #ìƒìœ„ 15ê°œ í•­ëª©ë§Œ ê°€ì ¸ì˜¤ê¸°
df2.columns =['menu', 'freq']
df2
```

```
# seaborn ë¼ì´ë¸ŒëŸ¬ë¦¬ barplotìœ¼ë¡œ í‘œí˜„!

import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt

plt.rc('font', family='NanumBarunGothic') 
import seaborn as sns

sns.barplot(
    data= df2,
    x= "menu",
    y= "freq"
)
plt.xticks(rotation = 45 )
```
